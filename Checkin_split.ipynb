```python
import pandas as pd
```

```python
Retrive_city = 'NewYork'
city_data = pd.read_csv('./dataset/NewYork_new.csv', encoding='ISO-8859-1')
city_data
```

```python
ind = city_data[city_data['Category']=='Home (private)'].index
```

```python
city_data=city_data.drop(ind)
city_data
```

```python
for num in range(1,57):
    print("File %d Loading......"%(num))
    data_checkin = pd.read_table('./Foursq_fileParts/dataset_TIST2015_Checkins_%d.txt'%(num),sep='\t',header=None) 
    data_checkin.columns = ['UserId','VenueId','Time','TimeZone']
    data_checkin
    merge = pd.merge(data_checkin, city_data, how='inner', on=['VenueId'])
    print(merge.shape)
    #print(merge)
    if num ==1:
        new_data = merge
        print(new_data.shape)
    else:
        frames = [new_data, merge]
        new_data = pd.concat(frames)
        print(new_data.shape)
print("Finished!")
```

```python
new_data
```

```python
new_data = new_data.drop_duplicates()
new_data
```

```python
new_data.to_csv("%s_checkin.csv"%(Retrive_city),index=False,sep=',')
print('Successful！')
```
