```python
distribution_data= pd.read_csv('./Distribution_Capare4City.csv', encoding='ISO-8859-1')
```

```python
distribution_data.index = pd.Series(['Phoenix', 'Calgary','NewYork','Singapore'])
distribution_data
```

```python
distribution_data = distribution_data.stack().unstack(0)
distribution_data
```

```python
distribution_data.loc[:,'Phoenix'].max()
```

```python
category_max_phoenix = distribution_data.loc[:,'Phoenix'].max()
category_max_calgary = distribution_data.loc[:,'Calgary'].max()
#print(category_max_calgary)
category_max_newyork = distribution_data.loc[:,'NewYork'].max()
category_max_singapore = distribution_data.loc[:,'Singapore'].max()
phoenix_similar_vec = []
calgary_similar_vec = []
newyork_similar_vec = []
singapore_similar_vec = []
for index, row in distribution_data.iterrows():
    num =0
    if row[0] != 0:
        num = num+1
    if row[1] != 0:
        num = num+1
    if row[2] != 0:
        num = num+1
    if row[3] != 0:
        num = num+1
    IDF = np.log(4/(num+1))
    #Phoenix
    TF_phoenix = row[0]/category_max_phoenix
    phoenix_similar_vec.append(TF_phoenix*IDF)
    ##calgary
    TF_calgary = row[1]/category_max_calgary
    calgary_similar_vec.append(TF_calgary*IDF)
    #newyork
    TF_newyork = row[2]/category_max_newyork
    newyork_similar_vec.append(TF_newyork*IDF)
    #singapore
    TF_singapore = row[3]/category_max_singapore
    singapore_similar_vec.append(TF_singapore*IDF)
```

```python
def list_dot_plus(list1,list2):
    sum =0
    for i in range(len(list1)):
        a = list1[i] * list2[i]
        sum =  a+sum
    return sum 
```
