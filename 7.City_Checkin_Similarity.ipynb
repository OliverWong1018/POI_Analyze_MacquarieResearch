```python
import pandas as pd
```

```python
checkin_phoenix = pd.read_csv('./dataset/Phoenix_c2c_count.csv', encoding='ISO-8859-1')
checkin_calgary = pd.read_csv('./dataset/Calgary_c2c_count.csv', encoding='ISO-8859-1')
checkin_newyork = pd.read_csv('./dataset/NewYork_c2c_count.csv', encoding='ISO-8859-1')
checkin_singapore= pd.read_csv('./dataset/Singapore_c2c_count.csv', encoding='ISO-8859-1')
print(checkin_phoenix.shape)
print(checkin_calgary.shape)
print(checkin_newyork.shape)
print(checkin_singapore.shape)
```

```python
data= pd.DataFrame()
data = data.append(checkin_phoenix,ignore_index= True)
data = data.append(checkin_calgary,ignore_index= True)
data = data.append(checkin_newyork,ignore_index= True)
data = data.append(checkin_singapore,ignore_index= True)
data
```

```python
data=data.fillna(0)
data.astype(int)
```

```python
data.index = pd.Series(['Phoenix', 'Calgary','NewYork','Singapore'])
data.to_csv("Checkin_Capare4City.csv",index=False,sep=',')
data
```

```python
data = data.stack().unstack(0)
data
```

```python
data['Phoenix'] = data['Phoenix'].div(data['Phoenix'].sum())
data['Calgary'] = data['Calgary'].div(data['Calgary'].sum())
data['NewYork'] = data['NewYork'].div(data['NewYork'].sum())
data['Singapore'] = data['Singapore'].div(data['Singapore'].sum())
data
```

```python
from scipy.spatial.distance import cosine
```

```python
pho = data['Phoenix'].values.tolist()
cal = data['Calgary'].values.tolist()
new = data['NewYork'].values.tolist()
sin = data['Singapore'].values.tolist()
```

```python
print(cosine(pho, cal))
print(cosine(pho, new))
print(cosine(pho, sin))
print(cosine(cal, new))
print(cosine(cal, sin))
print(cosine(new, sin))
```

```python
def list_dot_plus(list1,list2):
    sum =0
    for i in range(len(list1)):
        a = list1[i] * list2[i]
        sum =  a+sum
    return sum 
```

```python
print(list_dot_plus(pho, cal))
print(list_dot_plus(pho, new))
print(list_dot_plus(pho, sin))
print(list_dot_plus(cal, new))
print(list_dot_plus(cal, sin))
print(list_dot_plus(new, sin))
```
